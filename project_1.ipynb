# Research methods
- Data resources: Ministry of Transportation and Communications, ROC
- Website link: https://stat.motc.gov.tw/mocdb/stmain.jsp?sys=100&funid=emenu
- Obtained three different tables respectively for Taoyuan MRT, Taoyuan International Airport and Taipei MRT.
- Observed how the pandemic affected international travel at Taiwan's main Aviation hub, Taoyuan International Airport.
- Observed the traffic condition of Taoyuan Metro and compared it with Taipei metro. 
- Drew four figures to represent the traffic conditions of Taoyuan International Airport, Taoyuan Metro and Taipei Metro.






```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (18,6)
```

```python
airport_data = pd.read_csv("/content/drive/MyDrive/Colab_Notebooks/project_1/taoyuanAirport.csv", header= 0, encoding= 'unicode_escape')
airport_data
```

```python
def to_num(data): 
  lst= []
  for num in data:
    new= int(num.replace(",",""))
    lst.append(new)
    arr = np.array(lst)
  return arr
```

```python
airport_data_X = np.array(airport_data["Year/month"])
airport_data_Y = to_num(np.array(airport_data["pessenger(k)"]))
airport_data_Y_arrival = to_num(np.array(airport_data["arrival(k)"]))
airport_data_Y_departure = to_num(np.array(airport_data["departure(k)"]))
airport_data_Y_last_year = np.array(airport_data["last_year(%)"])
print(airport_data_Y_last_year)
```

```python
tyMrt_data = pd.read_csv("/content/drive/MyDrive/Colab_Notebooks/project_1/taoyuanMRT.csv")
tyMrt_data.head()
```

```python
tyMrt_data_X = np.array(tyMrt_data["Year/month"])
tyMrt_data_Y = np.array(tyMrt_data["pessenger(k)"])
tyMrt_data_Y_last_year = np.array(tyMrt_data["last_year(%)"])
# print(tyMrt_data_X)

tyMrt_data_Y = to_num(tyMrt_data_Y)
print(tyMrt_data_Y_last_year)
```

```python
tpMrt_data = pd.read_csv("/content/drive/MyDrive/Colab_Notebooks/project_1/taipeiMRT.csv")
tpMrt_data.head()
```

```python
tpMrt_data_X = np.array(tpMrt_data["Year/month"])
tpMrt_data_Y = to_num(np.array(tpMrt_data["pessenger(k)"]))
tpMrt_data_Y_last_year = np.array(tpMrt_data["last_year(%)"])
print(tpMrt_data_Y_last_year)
```

```python
plt.plot(airport_data_X ,airport_data_Y, color="indianred", label="Total visitors")
plt.plot(airport_data_X ,airport_data_Y_arrival, color="violet", label="Arrival visitors")
plt.plot(airport_data_X ,airport_data_Y_departure, color="turquoise", label="Departure visitors")
plt.legend(loc="upper right")
plt.title("Fig 1. Passenger Number Per Month of Taoyuan International Airport in Recent 5 Years")
plt.xlabel("Time (Year/Month)")
plt.ylabel("Passengers(1,000 Passengers)")
plt.tick_params(axis='x', colors='black', direction='out', length=10, width=1)
plt.gcf().autofmt_xdate()
plt.xticks(fontsize=9)
plt.grid(True)
plt.show()
```

```python
plt.plot(tyMrt_data_X ,tyMrt_data_Y ,color="steelblue")
plt.title("Fig 2. Passenger Number Per Month of Taoyuan MRT in Recent 5 Years")
plt.xlabel("Time (Year/Month)")
plt.ylabel("Passengers(1,000 Passengers)")
plt.tick_params(axis='x', colors='black', direction='out', length=10, width=1)
plt.gcf().autofmt_xdate()
plt.xticks(fontsize=9)
plt.grid(True)
plt.show()
```

```python
plt.plot(tpMrt_data_X  ,tpMrt_data_Y ,color="forestgreen")
plt.title("Fig 3. Passenger Number Per Month of Taipei MRT in Recent 5 Years")
plt.xlabel("Time (Year/Month)")
plt.ylabel("Passengers(1,000 Passengers)")
plt.tick_params(axis='x', colors='black', direction='out', length=10, width=1)
plt.gcf().autofmt_xdate()
plt.xticks(fontsize=9)
plt.grid(True)
plt.show()
```

```python
plt.plot(airport_data_X ,airport_data_Y_last_year, color="indianred", label="Taoyuan International Airpot")
plt.plot(tyMrt_data_X ,tyMrt_data_Y_last_year, color="steelblue", label="Taoyuan MRT")
plt.plot(tpMrt_data_X ,tpMrt_data_Y_last_year, color="forestgreen", label="Taipei MRT")
plt.legend(loc="upper left")
plt.title("Fig 4. Percentage Changes of Passenger Number Compared with Same Period Last Year")
plt.xlabel("Time (Year/Month)")
plt.ylabel("Change percentage (%)")
plt.tick_params(axis='x', colors='black', direction='out', length=10, width=1)
plt.gcf().autofmt_xdate()
plt.xticks(fontsize=9)
plt.grid(True)
plt.show()
```

# Conclusions
1. The traffic of Taoyuan International Airport was affected most severely by the pandemic. The passenger number dropped for around 100% within the first four months since the pandemic outbreak in early 2020 and remained the low number since then due to the border policy. 
2. The passenger number of Taoyuan International Airport slowly increased for the past three months ( from May 2022) after the border policy was changed. 
3. The upward sign at the end of the plot was observed during May 2022 and June 2022. 
4. On the figures of Taoyuan Metro and Taipei Metro passenger numbers per month, two obvious valleys were noted. The bottom points were located in April 2020 and June 2021. 
5. The Taoyuan metro traffic was not only affected by international travel but also closely connected with the national pandemic situation. The two figures revealed the same pattern of passenger numbers after the Covid-19 pandemic. 
6. Taoyuan metro offers a fast way of commuting between Taoyuan Airport and downtown Taipei. It also plays an important role in the transportation between Taoyuan City and the Great Taipei area.
